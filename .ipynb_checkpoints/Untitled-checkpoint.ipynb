```python
from selenium import webdriver
import time
driver = webdriver.Chrome()
driver.get('https://phukiengiare.com/bao-da-op-lung-phu-kien-samsung.html')
```

```python
# function
def showMoreOption():
    while(True):
        time.sleep(3)
        try:
            driver.find_element_by_css_selector('a.more_load_page').click()
            print("true")
        except:
            print("done")
            break
```

```python
# main
```

```python
# test

showMoreOption()
container = driver.find_element_by_css_selector('div.grid_product_list_column.clearfix')
```

```python
listItem = container.find_elements_by_css_selector('div.product_list_column5')
```

```python
listDic = []
for i in listItem:
    dic = {}
    try:
        dic["title"] = i.find_element_by_css_selector('div.grid-list__item-name>a').text
    except:
        dic["title"] = "None"
    try:
        dic["price"] = i.find_element_by_css_selector('span.price').text
    except:
        dic["price"] = "Unknow"
    try:
        dic["image"] = i.find_element_by_css_selector('img.pict').get_attribute('src')
    except:
        dic["image"] = "Unkown"
    listDic.append(dic)
    
listDic
```
